```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Baca citra asli menggunakan OpenCV
image = cv2.imread('Presiden_Sukarno.jpg', cv2.IMREAD_GRAYSCALE)  # Pastikan untuk mengganti path sesuai dengan lokasi citra Anda

# Validasi apakah citra berhasil dibaca
if image is None:
    print("Error: Gagal membaca citra. Pastikan path sudah benar.")
    exit()

# Hitung histogram dari citra asli
histogram = cv2.calcHist([image], [0], None, [256], [0, 256])

# Lakukan histogram equalization pada citra
equalized_image = cv2.equalizeHist(image)

# Hitung histogram dari citra setelah histogram equalization
histogram_equalized = cv2.calcHist([equalized_image], [0], None, [256], [0, 256])

# Tampilkan citra asli dan citra hasil histogram equalization
tl = plt.figure(figsize=(10, 10))
plt.subplot(2, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Citra Asli')
plt.axis('off')

# Tampilkan histogram citra asli
plt.subplot(2, 2, 2)
plt.plot(histogram, color='black')
plt.title('Histogram Citra Asli')
plt.xlabel('Intensitas Pixel')
plt.ylabel('Jumlah Pixel')

# Tampilkan citra hasil histogram equalization
plt.subplot(2, 2, 3)
plt.imshow(equalized_image, cmap='gray')
plt.title('Citra Setelah Histogram Equalization')
plt.axis('off')

# Tampilkan histogram citra setelah histogram equalization
plt.subplot(2, 2, 4)
plt.plot(histogram_equalized, color='black')
plt.title('Histogram Citra Setelah Histogram Equalization')
plt.xlabel('Intensitas Pixel')
plt.ylabel('Jumlah Pixel')

plt.tight_layout()
plt.show()
```
